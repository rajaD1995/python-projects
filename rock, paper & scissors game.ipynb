rock, paper, scissors game:
1. rock > scissors: scissors > paper: paper > rock: condition to win
2. The game will goes on until user stop it.
3. user can input only numbers so error handling is also covered.

```python
import random

# ASCII Art for rock, paper, scissors
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

# Choices list (0: Rock, 1: Paper, 2: Scissors)
choice_list = [rock, paper, scissors]

k = True  # Loop control variable

while k:
    try:
        # Get user input
        user_input = int(input("Please enter 0 for Rock, 1 for Paper, 2 for Scissors: "))

        # Validate input
        if 0 <= user_input < 3:
            # Generate system's choice for each round
            system_input = random.randint(0, 2)

            print(f"You chose:\n{choice_list[user_input]}")
            print(f"System chose:\n{choice_list[system_input]}")

            # Determine the winner
            if (user_input == 0 and system_input == 2) or (user_input == 1 and system_input == 0) or (user_input == 2 and system_input == 1):
                print("ðŸŽ‰ Congratulations! You won!")
            elif user_input == system_input:
                print("ðŸ˜² It's a draw!")
            else:
                print("ðŸ˜¢ Sorry! You lost!")

            # Ask user if they want to continue
            user_choice = input("Do you want to continue? (Y or N): ").strip().upper()
            if user_choice == 'N':
                k = False  # Exit the loop

        else:
            print("âš ï¸ Please select a valid input (0, 1, or 2).")

    except ValueError:
        print("âš ï¸ Invalid input! Please enter a number (0, 1, or 2).")

print("Thanks for playing! Goodbye! ðŸ‘‹")
```
